```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#
# Transformación del dataset para facilitar el análisis
#
iris = sns.load_dataset("iris")

iris_melt = iris.copy()
iris_melt = pd.melt(
    iris_melt,
    id_vars="species",
    var_name="Variables",
    value_name="Values",
)
iris_melt
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
plt.show()
```

```python
#
# Relación entre el largo y el ancho del petalo
#
sns.lmplot(
    x="petal_width",
    y="petal_length",
    hue="species",
    data=iris,
    palette="Set1",
)

plt.show()
```
