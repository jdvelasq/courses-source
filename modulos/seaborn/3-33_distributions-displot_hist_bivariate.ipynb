Histogramas bivariados con displot() ---
===

* 0:00 min | Última modificación: Octubre 13, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
#
# Heatmap básico
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    cbar=True,
    rug=True,
    kind='hist',
)
plt.show()
```

```python
#
# Heatmap por variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    hue="species",
    rug=True,
    kind='hist',    
)
plt.show()
```
