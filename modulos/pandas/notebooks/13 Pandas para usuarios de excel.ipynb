## Subtotales

```python
#
# Contar -- size()
#
tips.groupby("sex").size()
```

```python
#
# Contar -- size() vs count()
#
tips.groupby("sex").count()
```

```python
#
# Contar -- max(), min(), ...
#
tips.groupby("sex").max()
```

```python
#
# Funciones diferenciales por columna
#
tips.groupby("sex").agg(
    {
        "tip": np.max,
        "total_bill": np.mean,
    }
)
```

```python
tips.groupby(
    "sex",
    as_index=True,
).agg(
    {
        "tip": np.max,
        "total_bill": np.mean,
    }
)
```

```python
#
# as_index = False
#
tips.groupby(
    "sex",
    as_index=False,
).agg(
    {
        "tip": np.max,
        "total_bill": np.mean,
    }
)
```

## Tablas dinámicas

```python
#
# Del ejemplo anterior
#
def f(a, b, c):
    return a + b + c


escenarios = [
    {"a": 1, "b": 2, "c": 3},
    {"a": 4, "b": 5, "c": 6},
    {"a": 7, "b": 8, "c": 9},
    {"a": 10, "b": 11, "c": 12},
]

z = pd.DataFrame(range(len(escenarios)), columns=["Escenario"])
z
```

```python
z["f"] = [f(**escenarios[i]) for i in z["Escenario"]]
z
```

## Solver

```python
def f0(x):
    return sum([i**2 for i in x])


# 1^2 + 2^2 + 3^2 = 14
f0([1, 2, 3])
```

```python
from scipy.optimize import minimize

minimize(f0, x0=[1, 2, 3]).x
```

## Styles

Basado en:

https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html

```python
import numpy as np

np.random.seed(24)
df = pd.DataFrame({"A": np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list("BCDE"))], axis=1)
df.iloc[3, 3] = np.nan
df.iloc[0, 2] = np.nan

df
```

```python
def color_negative_red(val):
    color = "red" if val < 0 else "black"
    return "color: %s" % color


df.style.applymap(color_negative_red)
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ["background-color: yellow" if v else "" for v in is_max]


df.style.apply(highlight_max)
```

```python
df.style.applymap(color_negative_red).apply(highlight_max)
```

```python
df.style.apply(highlight_max, subset=["B", "C", "D"])
```

```python
df.style.applymap(
    color_negative_red,
    subset=pd.IndexSlice[2:5, ["B", "D"]],
)
```

```python
df.style.format("{:.2%}")
```

```python
df.style.format(
    {
        "B": "{:0<4.0f}",
        "D": "{:+.2f}",
    }
)
```

```python
df.style.format(
    {
        "B": lambda x: "±{:.2f}".format(abs(x)),
    }
)
```

```python
df.style.format(
    "{:.2%}",
    na_rep="-",
)
```

```python
df.style.highlight_null(color="red")
```

```python
import seaborn as sns

cm = sns.light_palette("green", as_cmap=True)

df.style.background_gradient(cmap=cm)
```

```python
df.loc[:4].style.background_gradient(cmap="viridis")
```

```python
df.style.highlight_max(axis=0)
```

```python
df.style.bar(
    subset=["A", "B"],
    color="#d65f5f",
)
```

```python
df.style.bar(
    subset=["A", "B"],
    align="mid",
    color=["#d65f5f", "#5fba7d"],
)
```
