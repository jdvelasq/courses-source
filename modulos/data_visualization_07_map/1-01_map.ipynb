```python
import matplotlib.pyplot as plt
import geopandas
import geoplot

path = geopandas.datasets.get_path('naturalearth_lowres')
df = geopandas.read_file(path)
df['gdp_pp'] = df.gdp_md_est / df.pop_est

df.head()
```

```python
geoplot.choropleth(
    df,
    hue="gdp_pp",
    cmap="Blues",
    figsize=(12, 6),
)

plt.show()
```
