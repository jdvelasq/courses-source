Tipos de línea y marcadores
===

*Diplomado en Analítica para Mercados de Energía*

A continuación se presentan los principales tipos de gráficos para la visualización de información, los cuales esán disponibles en la mayoría de aplicaciones.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 10)
```

Estilos de línea
---

```python
linestyles = ["solid", "dashed", "dashdot", "dotted"]

for i_style, style in enumerate(linestyles):
    plt.plot(x, x + i_style, linestyle=style, label=style)

plt.legend()
plt.show()
```

```python
linestyles = ["-", "--", "-.", ":"]

for i_style, style in enumerate(linestyles):
    plt.plot(x, x + i_style, linestyle=style, label=style)

plt.legend()
plt.show()
```

Marcadores
---

```python
from matplotlib.lines import Line2D

plt.figure(figsize=(4, 14))

markers = [m for m, func in Line2D.markers.items()]

text_style = dict(
    horizontalalignment="right",
    verticalalignment="center",
    fontsize=14,
    fontfamily="monospace",
)

marker_style = dict(
    linestyle=":",
    color="0.8",
    markersize=12,
    markerfacecolor="tab:blue",
    markeredgecolor="tab:blue",
)

for y, marker in enumerate(markers):

    plt.text(-0.5, y, repr(marker), **text_style)
    plt.plot([y] * 3, marker=marker, **marker_style)
    # plt.gca().margins(0.2)
    plt.gca().set_axis_off()
    plt.gca().invert_yaxis()

plt.show()
```

```python
y = np.power(x, 2)

plt.plot(x, y, linestyle="--", marker="o")
plt.show()
```
