Gráficos estadísticos: pie(x)
===

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
smoker = tips.smoker.value_counts() / len(tips) * 100
smoker
```

## Gráfico circular

```python
plt.figure(figsize=(3.5, 3.5))

colors=['orange', 'mediumblue']

wedges = plt.pie(
    x=smoker,
    explode=[0.1, 0],
    labels=smoker.index.tolist(),
    colors=colors,
    shadow=False,
    startangle=90,
    autopct="%1.1f%%",
    counterclock=True,
)

#
# Formato de las tortas
#
for i, w in enumerate(wedges[0]):
    w.set_edgecolor(colors[i])
    w.set_linewidth(3)
    w.set_alpha(0.3)
    
plt.gca().axis("equal")
plt.show()
```

## Donut Plot

```python
plt.figure(figsize=(3.5, 3.5))

colors = ["orange", "mediumblue"]
plt.pie(
    x=smoker,
    labels=smoker.index.tolist(),
    colors=colors,
    shadow=False,
    startangle=90,
    counterclock=True,
    wedgeprops={"linewidth": 10, "edgecolor": "white"},
)

my_circle = plt.Circle((0, 0), 0.7, color="white")
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.gca().axis("equal")
plt.show()
```
