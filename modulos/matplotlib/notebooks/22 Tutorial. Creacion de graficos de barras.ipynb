Caso de estudio: Creación de gráficos de barra
===

```python
import matplotlib.pyplot as plt

import seaborn as sns
```

Carga de datos
---

```python
tips = sns.load_dataset("tips")
smoker = tips.smoker.value_counts()
smoker
```

Barras verticales
---

```python
plt.figure(figsize=(5, 4))

container = plt.bar(
    x=smoker.index,
    height=smoker,
    color="tab:blue",
    alpha=0.5,
    edgecolor="tab:blue",
    linewidth=3,
)

for i_text, (text, value) in enumerate(zip(smoker.index, smoker)):
    plt.text(
        i_text,
        value / 2,
        value,
        ha="center",
        va="center",
        size=20,
    )

plt.title("Number of bills", fontsize=18)
plt.xlabel("Smoker", fontsize=14)
plt.ylabel("Quantity")


plt.gca().spines["left"].set_visible(False)
plt.gca().spines["bottom"].set_color("gray")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.show()
```

Barras horizontales
---

```python
plt.figure(figsize=(4, 3))

plt.barh(
    y=range(len(smoker)),
    width=smoker,
    color="tab:green",
    alpha=0.5,
    edgecolor="tab:green",
    linewidth=3,
)

plt.title("Number of bills", fontsize=18)
plt.xlabel("Smoker", fontsize=14)
plt.ylabel("Quantity")

plt.yticks(
    ticks=range(len(smoker)),
    labels=smoker.index,
    rotation="horizontal",
)

plt.gca().spines["left"].set_color("gray")
plt.gca().spines["bottom"].set_color("gray")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```

Barras apiladas
---

```python
smoker_sex = tips.groupby(["smoker", "sex"]).size()

smoker_sex_yes = smoker_sex["Yes"]
smoker_sex_no = smoker_sex["No"]

display(
    smoker_sex_yes,
    smoker_sex_no,
)
```

```python
plt.figure(figsize=(6, 3))

plt.bar(
    x=smoker_sex_yes.index,
    height=smoker_sex_yes,
    color="tab:blue",
    alpha=0.5,
    edgecolor="tab:blue",
    linewidth=3,
    label="Yes",
)

plt.bar(
    x=smoker_sex_no.index,
    height=smoker_sex_no,
    bottom=smoker_sex_yes,
    color="tab:orange",
    alpha=0.5,
    edgecolor="tab:orange",
    linewidth=3,
    label="No",
)

plt.title("Number of bills", fontsize=18)
plt.xlabel("Smoker", fontsize=14)
plt.ylabel("Quantity")

plt.xticks(
    rotation="vertical",
)

plt.legend()

plt.gca().spines["left"].set_color("lightgray")
plt.gca().spines["bottom"].set_color("gray")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```
