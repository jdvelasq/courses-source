```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights_long = sns.load_dataset("flights")
flights_long.head()
```

```python
flights = flights_long.pivot(
    index="month",
    columns="year",
    values="passengers",
)
flights
```

```python
#
# Figura b√°sica
#
fig, ax = plt.subplots(figsize=(8,3.5))
ax.pcolormesh(
    flights.columns,
    flights.index,
    flights.values,
)
plt.show()
```
