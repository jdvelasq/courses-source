Gráficos básicos: imshow(Z)
===

```python
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(
    np.linspace(-3, 3, 16),
    np.linspace(-3, 3, 16),
)
Z = (1 - X / 2 + X**5 + Y**3) * np.exp(-(X**2) - Y**2)

fig, ax = plt.subplots(figsize=(3.5, 3.5))
ax.imshow(Z)
plt.show()
```
