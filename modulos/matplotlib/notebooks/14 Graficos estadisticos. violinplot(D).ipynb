Gráficos estadísticos: violinplot(D)
===

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(10)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))
df = pd.DataFrame(D, columns=["A", "B", "C"])

fig, ax = plt.subplots(figsize=(8, 3.5))

vp = ax.violinplot(
    df,
    [2, 4, 6],
    widths=2,
    showmeans=True,
    showmedians=False,
    showextrema=False,
)

for body in vp["bodies"]:
    body.set_alpha(0.6)

ax.set(
    xlim=(0, 8),
    xticks=np.arange(1, 8),
    ylim=(0, 8),
    yticks=np.arange(1, 8),
)

plt.show()
```
