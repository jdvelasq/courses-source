Gráficos básicos: evenplot(x, y)
===

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(1)
x = [2, 4, 6]
D = np.random.gamma(4, size=(3, 50))
D
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5))

ax.eventplot(
    D,
    orientation="vertical",
    lineoffsets=x,
    linewidth=0.75,
)

plt.show()
```
