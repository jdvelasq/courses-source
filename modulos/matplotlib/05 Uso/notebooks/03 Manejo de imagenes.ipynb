Tutorial básico de manejo de imágenes
===

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
#
# Una imagen RGB PNG es representada coomo un array de 3 dimensiones que
# indican los colornes [R, G, B] de cada pixel.
img = np.asarray(Image.open("assets/stinkbug.png"))

# Dimensiones de la imagen
print(img.shape)

# Representación como un array de NumPy
print(repr(img))
```

```python
img.shape
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(img)
plt.show()
```

```python
#
# Aplicación de un seudocolor para mejorar el contraste
#
plt.figure(figsize=(4, 4))
lum_img = img[:, :, 0]
plt.imshow(lum_img)
plt.show()
```

```python
#
# Aplicación de una luminosidad
#
plt.figure(figsize=(4, 4))
plt.imshow(lum_img, cmap="hot")
plt.show()
```

```python
#
# Cambio de colormap de un objeto existente
#
plt.figure(figsize=(4, 4))
imgplot = plt.imshow(lum_img)
imgplot.set_cmap("nipy_spectral")
```

```python
#
# Escala de color de referencia
#
plt.figure(figsize=(4, 4))
imgplot = plt.imshow(lum_img)
plt.colorbar()
plt.show()
```

```python
#
# Examinación de un rango de datos
#
plt.figure(figsize=(4, 3.5))
plt.hist(lum_img.ravel(), bins=range(256), fc="k", ec="k")
plt.show()
```

```python
#
# Se resalta la información del rango del pico
#
plt.figure(figsize=(4, 4))
plt.imshow(lum_img, clim=(0, 175))
plt.show()
```

```python
#
# Esta misma tarea puede ser realizada con set_clim()
#
plt.figure(figsize=(4, 4))
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0, 175)
plt.show()
```

```python
#
# Redimensionamiento de la imagen e interpolacion (nearest)
#
plt.figure(figsize=(4, 4))
img = Image.open("assets/stinkbug.png")
img.thumbnail((64, 64))
imgplot = plt.imshow(img)
```

```python
#
# Interpolación bilineal
#
plt.figure(figsize=(4, 4))
imgplot = plt.imshow(img, interpolation="bilinear")
```

```python
#
# Interpolación bicubica
#
plt.figure(figsize=(4, 4))
imgplot = plt.imshow(img, interpolation="bicubic")
```
