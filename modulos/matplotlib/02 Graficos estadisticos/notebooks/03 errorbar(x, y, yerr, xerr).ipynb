Gráficos estadísticos: errorbar(x, y, yerr, xerr)
===

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 1.2, 0.5]

fig, ax = plt.subplots(figsize=(6, 3.5))
ax.errorbar(
    x,
    y,
    yerr,
    fmt="o",
    linewidth=2,
    capsize=8,
)
ax.set(
    xlim=(0, 8),
    xticks=np.arange(1, 8),
    ylim=(0, 8),
    yticks=np.arange(1, 8),
)

plt.show()
```
