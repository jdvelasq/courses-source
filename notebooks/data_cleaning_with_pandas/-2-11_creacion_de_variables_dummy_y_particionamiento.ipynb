## Preparacion

```python
import pandas as pd
```

## Carga de datos

```python
iris_flowers = pd.read_csv(
    "https://raw.githubusercontent.com/jdvelasq/datalabs/master/datasets/iris.csv",
    sep = ',',         
    thousands = None,  
    decimal = '.')     

iris_flowers.head()
```

## Creación de variables dummy

```python
iris_flowers_dummies = pd.concat(
    [
        iris_flowers,
        pd.get_dummies(iris_flowers.Species),
    ],
    axis = 1,
)
iris_flowers_dummies
```

## Mezcla de dataset

```python
iris_flowers_dummies = iris_flowers_dummies.sample(frac = 1)
iris_flowers_dummies
```

## Subconjuntos de datos

```python
iris_flowers_dummies_train =  iris_flowers_dummies.sample(frac = 0.8)
iris_flowers_dummies_train
```

```python
iris_flowers_dummies_test = iris_flowers_dummies.drop(iris_flowers_dummies_train.index)
iris_flowers_dummies_test
```
