```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## sns.distplot()

```python
np.random.seed(0)
mu, sigma = 0, 1

data = np.random.normal(mu, sigma, 1000)

sns.distplot(data, color="#F15B2A", kde=True, bins=50,
             hist_kws=dict(edgecolor="#F15B2A", linewidth=2),
           kde_kws=dict(shade=True))
```

## sns.kdeplot()

```python
sns.kdeplot(data,shade=True, color='blue')
```

## sns.boxplot()

```python
p = sns.boxplot(
    data,
    palette=['#F15B2A'],
)

for e in p.artists:
    r, g, b, a = e.get_facecolor()
    p.set_facecolor((r, g, b, .0))
    
for i, artist in enumerate(p.artists):
    artist.set_edgecolor("#F15B2A")
    
for j in range(0, 6):
    line = p.lines[j]
    line.set_color("#F15B2A")
    line.set_mfc("#F15B2A")
    line.set_mec("#F15B2A")
```

## sns.violinplot()

```python
p = sns.violinplot(
    data,
    palette=['#F15B2A']
)
```

```python
#
# Se usa matplotlib para cambiar las propiedades
# de los elementos
#
parts = plt.violinplot(
    data,
    vert=False,
    showmeans=True,
    showmedians=True,
)

for p in parts['bodies']:
    p.set_facecolor('#F15B2A')
    p.set_edgecolor('#F15B2A')
    p.set_alpha(0.4)
```

## sns.stripplot()

```python
tips = sns.load_dataset('tips')
p = sns.stripplot(
    x=tips.total_bill,
    jitter=1,
)
```

## sns.swarmplot()

```python
p = sns.swarmplot(
    x=tips.total_bill,
)
```
