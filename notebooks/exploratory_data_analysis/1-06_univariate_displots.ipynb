```python
import warnings

warnings.filterwarnings("ignore")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris_melt = iris.copy()
iris_melt = pd.melt(
    iris_melt,
    id_vars="species",
    var_name="Variables",
    value_name="Values",
)
iris_melt.head()
```

```python
#
# Distribución de las variables
#
sns.catplot(
    x="Variables",
    y="Values",
    data=iris_melt,
    kind="swarm",
)
plt.show()
```

```python
#
# Histogramas
#
sns.displot(
    iris_melt,
    x="Values",
    col="species",
)
plt.show()
```

```python
iris_melt = iris.copy()
iris_melt = pd.melt(
    iris_melt,
    id_vars="species",
    var_name="Variables",
    value_name="Values",
)
sns.catplot(
    x="Variables",
    y="Values",
    data=iris_melt,
    col="species",
    kind="swarm",
)
plt.show()
```

```python
#
# Histogramas
#
sns.displot(
    iris_melt,
    x="Values",
    col="Variables",
    row='species',
    stat="density",
    bins=20,
)
plt.show()
```

```python
#
# Estimación empírica de la densidad de probabilidad
#
sns.displot(
    iris_melt,
    x="Values",
    col="Variables",
    hue='species',
    kind='kde',
    fill=True,
)
plt.show()
```

```python
#
# Densidad acumulada de probabilidad empírica
#
sns.displot(
    iris_melt,
    x="Values",
    col="Variables",
    hue='species',
    kind='ecdf',
)
plt.show()
```

```python
sns.catplot(
    x="species",
    y='Values',
    row='Variables',
    data=iris_melt,
    kind='box',
)
plt.show()
```
