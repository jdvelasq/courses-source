```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color="#F15B2A")
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', color="#F15B2A")
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(tips['total_bill'], tips['tip'], tips['size'])
```

```python
ax.bar3d(tips['total_bill'], tips['tip'], np.zeros(len(tips['size'])), 1, 1, tips['size'])
```

```python
sns.boxplot(x='time', y='tip', hue='smoker', data=tips)
```

```python
sns.violinplot(x='time', y='tip', hue='smoker', data=tips)
```
