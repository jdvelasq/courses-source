```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

El [conjunto de datos de la flor Iris](https://es.wikipedia.org/wiki/Iris_flor_conjunto_de_datos) contiene 150 muestras de las medidas del ancho y el largo del pétalo y del sépalo para las tres especies de esta flor (Iris setosa, Iris virginica e Iris versicolor). 

```python
#
# Carga inicial de los datos
#
iris = sns.load_dataset("iris")
```

```python
#
# Visualización del dataset
#
iris.head()
```

```python
#
# Tipos de datos de las columnas
#
iris.info()
```

```python
#
# Conteo de casos para las variables categóricas
#
sns.countplot(x="species", data=iris)
plt.show()
```

```python
#
# Transformación del dataset para facilitar la visualización
#
iris_melt = iris.copy()
iris_melt = pd.melt(
    iris_melt,
    id_vars="species",
    var_name="Variables",
    value_name="Values",
)
iris_melt
```

```python
#
# Visualización preliminar de datos para las variables numéricas
#
sns.catplot(
    x="Variables",
    y="Values",
    data=iris_melt,
    kind="swarm",
)
plt.show()
```

```python
#
# Visualización preliminar por categoria
#
sns.catplot(
    x="Variables",
    y="Values",
    data=iris_melt,
    col="species",
    kind="swarm",
)
plt.show()
```
