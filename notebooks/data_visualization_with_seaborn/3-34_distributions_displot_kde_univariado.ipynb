KDE con displot() ---
===

* 0:00 min | Última modificación: Octubre 13, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")

display(
    penguins.head(),
    penguins.tail(),
    penguins.size,
)
```

```python
#
# Estimación no paramétrica básica usando dissplot(). 
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='tab:blue',
    fill=True,
    rug=True,
    cumulative=False,    
)
plt.show()
```

```python
#
# Modificación del ancho de banda para la estimación de los kernels.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=0.25,
)
plt.show()
```

```python
#
# Compare el resultado con el gráfico anterior.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=2,
)
plt.show()
```

```python
#
# kde por variable categórica.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    fill=True,
)
plt.show()
```

```python
#
# Efecto de apilamiento.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    multiple="stack",
)
plt.show()
```

```python
#
# kde acumulada usando displot().
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    cumulative=True,
)
plt.show()
```
