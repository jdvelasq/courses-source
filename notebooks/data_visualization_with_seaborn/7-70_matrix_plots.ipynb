```python
from string import ascii_letters

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#
# Heatmap básico
#
df = pd.DataFrame(
    np.random.random((10, 10)),
    columns=["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"],
)

sns.heatmap(
    df,
    annot=True,
    annot_kws={"size": 7},
)

plt.show()
```

```python
mtcars = pd.read_csv(
    "https://raw.githubusercontent.com/jdvelasq/datalabs/master/datasets/mtcars.csv",
)

mtcars = mtcars.set_index('model')

mtcars.head()
```

```python
sns.clustermap(
    mtcars, 
    metric="correlation", 
    method="single", 
    cmap="Blues", 
    standard_scale=1,
)

plt.show()
```
