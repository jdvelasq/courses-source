KDE bivariados con displot() ---
===

* 0:00 min | Última modificación: Octubre 13, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
#
# kde bivariado básico con especificación del
# número de curvas de nivel
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=5,
    rug=True,
)
plt.show()
```

```python
#
#
# kde bivariado básico con especificación de la
# ubicación de las curvas de nivel.
#
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=[.01, .05, .1, .8],
)
plt.show()
```

```python
#
# Separación por una variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)
plt.show()
```
