Series de tiempo en Seaborn
===

* 0:00 min | Última modificación: Octubre 13, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sutter = pd.read_csv(
    "https://raw.githubusercontent.com/jdvelasq/datalabs/master/datasets/sutter.csv",
)

#
# La columna time es leida como texto
# y no como fecha
#
display(
    sutter.head(),
    sutter.tail(),
    sutter.dtypes
)
```

```python
#
# Conversión de la la columna a datetime
#
sutter['time'] =  pd.to_datetime(sutter['time'], format='%Y-%m-%d')
display(
    sutter.head(),
    sutter.dtypes
)
```

```python
#
# Gráfico de la serie de tiempo
#
import matplotlib.dates as mdates

sns.set(rc={'figure.figsize':(12,6)})

sns.lineplot(
    x="time",
    y="value",
    data=sutter,
)

#
# Línea vertical
#
plt.axvline(
    pd.to_datetime('1964-01-01', format='%Y-%m-%d'),
    color='red',
    linestyle='--',
)
    
#
# Línea horizontal
#
plt.axhline(
    3000,
    color='green',
)

fmt_year = mdates.MonthLocator(bymonthday=1, interval=12)
plt.gca().xaxis.set_major_locator(fmt_year)
plt.gca().tick_params(axis="x", rotation=90)
plt.show()
```

```python
#
# Gráfico de la serie de tiempo
#
import matplotlib.dates as mdates

sns.set(rc={'figure.figsize':(12,6)})

sns.lineplot(
    x="time",
    y="value",
    data=sutter,
)

plt.axvspan(
    pd.to_datetime('1954-01-01', format='%Y-%m-%d'),
    pd.to_datetime('1957-12-01', format='%Y-%m-%d'),
    color='tab:gray',
    alpha=0.3,
)


plt.axhspan(
    3000,
    4000,
    color='tab:cyan',
    alpha=0.3,
)

fmt_year = mdates.MonthLocator(bymonthday=1, interval=12)
plt.gca().xaxis.set_major_locator(fmt_year)
plt.gca().tick_params(axis="x", rotation=90)
plt.show()
```
