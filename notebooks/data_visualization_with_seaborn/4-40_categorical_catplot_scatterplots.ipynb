Scatterplots categóricos usando catplot() ---
===

* 0:00 min | Última modificación: Octubre 12, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Stripplot()
---

```python
#
# Gráfico básico de puntos por categoría.
#
sns.catplot(
    x="day",
    y="total_bill",
    data=tips,
    kind='strip',
)

plt.show()
```

```python
#
# Efecto de jitter
#
sns.catplot(
    x="day",
    y="total_bill",
    jitter=False,
    data=tips,
    kind='strip',    
)

plt.show()
```

```python
#
# Gráfica para variables numéricas categóricas,
# modificando el orden del eje x.
#
sns.catplot(
    x="size",
    y="total_bill",
    data=tips,
    order=[2, 1, 3, 4, 5, 6],
    kind='strip',    
)

plt.show()
```

swarmplot()
---

```python
#
# Gráfica kind="swarm" con separación por
# categorías.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# La gráfica puede girarse con solo invertir las
# variables en los parámetros
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="time",
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# Separación por categorías
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="smoker",
    col="time",
    aspect=0.7,
    kind="swarm",
    data=tips,
)

plt.show()
```
