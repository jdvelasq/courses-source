Visualización de diferentes tipos de modelos con lmplot() ---
===

* 0:00 min | Última modificación: Octubre 13, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
anscombe = sns.load_dataset("anscombe")
tips = sns.load_dataset("tips")
```

```python
#
# Ajuste para uno de los subconjuntos de datos usando un modelo lineal
#
sns.lmplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'I'"),
    ci=None,
    scatter_kws={"s": 80},
)

plt.show()
```

```python
#
# Ajuste para el otro subconjunto de datos usando un modelo lineal
#
sns.lmplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'II'"),
    ci=None,
    scatter_kws={"s": 80},
)

plt.show()
```

```python
#
# AJuste usando un polinomio de grado 2.
#
sns.lmplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'II'"),
    order=2,
    ci=None,
    scatter_kws={"s": 80},
)

plt.show()
```

```python
#
# Ajuste en presencia de outliers.
#
sns.lmplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'III'"),
    ci=None,
    scatter_kws={"s": 80},
)

plt.show()
```

```python
#
# Regresión robusta. Requiere statsmodels.
#
sns.lmplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'III'"),
    robust=True,
    ci=None,
    scatter_kws={"s": 80},
)

plt.show()
```

```python
#
# Regresión logística.
#
tips["big_tip"] = (tips.tip / tips.total_bill) > .15

sns.lmplot(
    x="total_bill",
    y="big_tip",
    data=tips,
    logistic=True,
    y_jitter=0.03,
)

plt.show()
```

```python
#
# Regresión usando lowless
#
sns.lmplot(
    x="total_bill",
    y="tip",
    data=tips,
    lowess=True,
)

plt.show()
```

```python
#
# Gráfico de residuos del modelo
#
sns.residplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'I'"),
    scatter_kws={"s": 80},
)

plt.show()
```

```python
#
# Residuos con estructura
#
sns.residplot(
    x="x",
    y="y",
    data=anscombe.query("dataset == 'II'"),
    scatter_kws={"s": 80},
)

plt.show()
```
