```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Dataset

```python
penguins = sns.load_dataset("penguins")

display(
    penguins.head(),
    penguins.tail(),
    penguins.size,
)
```

## Densidad empírica

```python
#
# Estimación no paramétrica básica usando 
# dissplot(). 
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    rug=True,
    cumulative=False,    
)
plt.show()
```

```python
#
# Estimación no paramétrica básica usando 
# kdeplot(). Note que aca se pasa directamente
# la columna con los datos.
#
sns.kdeplot(
    penguins.flipper_length_mm,
    color="blue",
    shade=True,
)

plt.show()
```

```python
#
# kde acumulada usando displot().
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    cumulative=True,
)
plt.show()
```

```python
#
# kde acumulada usando kdeplot()
#
sns.kdeplot(
    penguins.flipper_length_mm,
    color="blue",
    shade=True,
    cumulative=True,
)

plt.show()
```

```python
#
# Modificación del ancho de banda para la 
# estimación de los kernels. Compare el 
# resultado con el gráfico siguiente.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=0.25,
)
plt.show()
```

```python
#
# Compare el resultado con el gráfico anterior.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=2,
)
plt.show()
```

```python
#
# kde por variable categórica.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    fill=True,
)
plt.show()
```

```python
#
# Efecto de apilamiento.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    multiple="stack",
)
plt.show()
```

## Distribuciones empíricas acumuladas

```python
#
# Gráfico básico.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="ecdf",
)
plt.show()
```

```python
#
# Gráfico básico por variable categórica
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="ecdf",
)
plt.show()
```

## Distribuciones bivariadas

```python
#
# Heatmap básico
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    cbar=True,
    rug=True,
)
plt.show()
```

```python
#
# Heatmap por variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    hue="species",
    rug=True,
)
plt.show()
```

```python
#
# kde bivariado básico con especificación del
# número de curvas de nivel
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=5,
    rug=True,
)
plt.show()
```

```python
#
#
# kde bivariado básico con especificación de la
# ubicación de las curvas de nivel.
#
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=[.01, .05, .1, .8],
)
plt.show()
```

```python
#
# Separación por una variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)
plt.show()
```

```python
#
# joinplot agrega las distribuciones marginales
# de las variables consideradas.
#
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)
plt.show()
```

```python
#
# Figura similar a la anterior, cambiando las
# gráficas a kde.
#
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)

plt.show()
```
