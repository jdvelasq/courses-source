```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#
# Gráfica básica de violin por categoría.
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="sex",
    kind="violin",
    data=tips,
)

plt.show()
```

```python
#
# Modificación de los parámetros.
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="sex",
    kind="violin",
    bw=0.15,
    cut=0,
    data=tips,
)

plt.show()
```

```python
#
# Construcción de una gráfica comparativa usando
# split.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="violin",
    split=True,
    data=tips,
)

plt.show()
```

```python
#
# Modificación de la visualización de los datos
# al interior de las figuras
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="violin",
    inner="stick",
    split=True,
    palette="pastel",
    data=tips,
)

plt.show()
```

```python
#
# Combinación de swarm con violin
#
g = sns.catplot(
    x="day",
    y="total_bill",
    kind="violin",
    inner=None,
    data=tips,
)

sns.swarmplot(
    x="day",
    y="total_bill",
    color="k",
    size=3,
    data=tips,
    ax=g.ax,
)

plt.show()
```
