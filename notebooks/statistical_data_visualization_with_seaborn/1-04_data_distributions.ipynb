```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Datasets utilizados

```python
penguins = sns.load_dataset("penguins")

display(
    penguins.head(),
    penguins.tail(),
    penguins.size,
)
```

```python
tips = sns.load_dataset("tips")

display(
    tips.head(),
    tips.tail(),
    tips.size,
)
```

## Graficación de histogramas univariados

```python
#
# Histograma básico con parámetros por defecto
# para una variable continua.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
)

plt.show()
```

```python
#
# Selección del ancho de cada barra para una
# variable continua.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    binwidth=3,
)

plt.show()
```

```python
#
# Especificación del número de barras para una
# variable continua.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    bins=20,
)

plt.show()
```

```python
#
# Histograma básico para una variable numérica
# discreta. Note que las barras no caen sobre los
# valores discretos que toma la variable.
#
sns.displot(
    tips,
    x="size",
)

plt.show()
```

```python
#
# Especificación de las barras
#
sns.displot(
    tips,
    x="size",
    bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5],
)

plt.show()
```

```python
#
# Se especifica directamente que la variable es
# numerica discreta.
#
sns.displot(
    tips,
    x="size",
    discrete=True,
)

plt.show()
```

```python
#
# Histograma para una Variable categórica. Se
# indica el ancho de las barras para que no se
# toquen.
#
sns.displot(
    tips,
    x="day",
    shrink=0.8,
)

plt.show()
```

## Histogramas por grupos

```python
#
# Separación de los histogramas por grupos. Note
# que van traslapados.
#
hue_colors = {
    'Gentoo': 'red',
    'Adelie': 'blue',
    'Chinstrap': 'yellow',
}

sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    hue_order=['Gentoo', 'Adelie', 'Chinstrap'],
    palette=hue_colors,
)

plt.show()
```

```python
#
# Se puede mejorar la visualización usando el
# parámetro element
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    element="step",
)
plt.show()
```

```python
#
# Histogramas apilados
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    multiple="stack",
)
plt.show()
```

```python
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="sex",
    multiple="dodge",
)
plt.show()
```

```python
sns.displot(
    penguins,
    x="flipper_length_mm",
    col="sex",
)
plt.show()
```

## sns.boxplot()

```python
#
# Box plot básico
#
sns.boxplot(
    x=penguins.flipper_length_mm,
    color='red',
)

plt.show()
```

```python
#
# Modificación de las propiedades
#
p = sns.boxplot(
    x=penguins.flipper_length_mm,
    color='lightblue',
)

for e in p.artists:
    r, g, b, a = e.get_facecolor()
    p.set_facecolor((r, g, b, 0.0))

for i, artist in enumerate(p.artists):
    artist.set_edgecolor("blue")

for j in range(0, 6):
    line = p.lines[j]
    line.set_color("red")
```

## sns.violinplot()

```python
#
# Violin plot
#
sns.violinplot(
    x=penguins.flipper_length_mm,
    palette=["lightblue"],
)
plt.show()
```

```python
#
# Se usa matplotlib para cambiar las propiedades
# de los elementos
#
parts = plt.violinplot(
    penguins.flipper_length_mm,
    vert=False,
    showmeans=True,
    showmedians=True,
)

for p in parts["bodies"]:
    p.set_facecolor("lightblue")
    p.set_edgecolor("blue")
    p.set_alpha(0.4)
```

## sns.swarmplot()

```python
sns.swarmplot(
    x=penguins.flipper_length_mm,
)
plt.show()
```

## Kernel density estimation

```python
#
# Estimación no paramétrica básica usando 
# dissplot(). 
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    rug=True,
)
plt.show()
```

```python
#
# Estimación no paramétrica básica usando 
# kdeplot(). Note que aca se pasa directamente
# la columna con los datos.
#
sns.kdeplot(
    penguins.flipper_length_mm,
    color="blue",
    shade=True,
)

plt.show()
```

```python
#
# kde acumulada usando displot().
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    cumulative=True,
)
plt.show()
```

```python
#
# kde acumulada usando kdeplot()
#
sns.kdeplot(
    penguins.flipper_length_mm,
    color="blue",
    shade=True,
    cumulative=True,
)

plt.show()
```

```python
#
# Modificación del ancho de banda para la 
# estimación de los kernels. Compare el 
# resultado con el gráfico siguiente.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=0.25,
)
plt.show()
```

```python
#
# Compare el resultado con el gráfico anterior.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=2,
)
plt.show()
```

```python
#
# kde por variable categórica.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    fill=True,
)
plt.show()
```

```python
#
# Efecto de apilamiento.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    multiple="stack",
)
plt.show()
```

## Visualización de distribuciones bivariadas

```python
#
# Heatmap básico
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    cbar=True,
    rug=True,
)
plt.show()
```

```python
#
# Heatmap por variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    hue="species",
    rug=True,
)
plt.show()
```

```python
#
# kde bivariado básico con especificación del
# número de curvas de nivel
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=5,
    rug=True,
)
plt.show()
```

```python
#
#
# kde bivariado básico con especificación de la
# ubicación de las curvas de nivel.
#
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=[.01, .05, .1, .8],
)
plt.show()
```

```python
#
# Separación por una variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)
plt.show()
```

## Gráficos compuestos

```python
#
# joinplot agrega las distribuciones marginales
# de las variables consideradas.
#
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)
plt.show()
```

```python
#
# Figura similar a la anterior, cambiando las
# gráficas a kde.
#
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)

plt.show()
```

```python
#
# Uso de JoinGrid para espeficiar el tipo de gráficas
# que se usaran en cada parte
#
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)

plt.show()
```

```python
#
# Combinación de dos gráficas
#
sns.relplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)

sns.rugplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)

plt.show()
```

## Graficación de muchas distribuciones simultáneamente

```python
#
# Gráfico usando pairplot()
#
sns.pairplot(penguins)
plt.show()
```

```python
#
# Uso de PairGrid() para especificar el tipo de
# gráfico en cada parte
#
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```
