```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")

display(
    penguins.head(),
    penguins.tail(),
    penguins.size,
)
```

```python
#
# Heatmap básico
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    cbar=True,
    rug=True,
)
plt.show()
```

```python
#
# Heatmap por variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    binwidth=(2, 0.5),
    hue="species",
    rug=True,
)
plt.show()
```

```python
#
# kde bivariado básico con especificación del
# número de curvas de nivel
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=5,
    rug=True,
)
plt.show()
```

```python
#
#
# kde bivariado básico con especificación de la
# ubicación de las curvas de nivel.
#
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    kind="kde",
    thresh=0.2,
    levels=[.01, .05, .1, .8],
)
plt.show()
```

```python
#
# Separación por una variable categórica.
#
sns.displot(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)
plt.show()
```

## Gráficos compuestos

```python
#
# joinplot agrega las distribuciones marginales
# de las variables consideradas.
#
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)
plt.show()
```

```python
#
# Figura similar a la anterior, cambiando las
# gráficas a kde.
#
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    kind="kde",
)

plt.show()
```

```python
#
# Uso de JoinGrid para espeficiar el tipo de gráficas
# que se usaran en cada parte
#
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)

plt.show()
```

```python
#
# Combinación de dos gráficas
#
sns.relplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)

sns.rugplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
)

plt.show()
```
