```python
#
# Gráfico básico de puntos por categoría.
#
sns.catplot(
    x="day",
    y="total_bill",
    data=tips,
)

plt.show()
```

```python
#
# Efecto de jitter
#
sns.catplot(
    x="day",
    y="total_bill",
    jitter=False,
    data=tips,
)

plt.show()
```

```python
#
# Gráfica kind="swarm" con separación por
# categorías.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# Gráfica para variables numéricas categóricas,
# modificando el orden del eje x.
#
sns.catplot(
    x="size",
    y="total_bill",
    data=tips,
    order=[2, 1, 3, 4, 5, 6],
)

plt.show()
```

```python
#
# La gráfica puede girarse con solo invertir las
# variables en los parámetros
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="time",
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# Separación por categorías
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="smoker",
    col="time",
    aspect=0.7,
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# Gráfica básica de violin por categoría.
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="sex",
    kind="violin",
    data=tips,
)

plt.show()
```

```python
#
# Modificación de los parámetros.
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="sex",
    kind="violin",
    bw=0.15,
    cut=0,
    data=tips,
)

plt.show()
```

```python
#
# Construcción de una gráfica comparativa usando
# split.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="violin",
    split=True,
    data=tips,
)

plt.show()
```

```python
#
# Modificación de la visualización de los datos
# al interior de las figuras
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="violin",
    inner="stick",
    split=True,
    palette="pastel",
    data=tips,
)

plt.show()
```

```python
#
# Combinación de swarm con violin
#
g = sns.catplot(
    x="day",
    y="total_bill",
    kind="violin",
    inner=None,
    data=tips,
)

sns.swarmplot(
    x="day",
    y="total_bill",
    color="k",
    size=3,
    data=tips,
    ax=g.ax,
)

plt.show()
```
