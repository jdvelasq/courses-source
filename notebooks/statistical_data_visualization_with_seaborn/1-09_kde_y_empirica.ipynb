```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")

display(
    penguins.head(),
    penguins.tail(),
    penguins.size,
)
```

```python
#
# Estimación no paramétrica básica usando 
# dissplot(). 
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    rug=True,
)
plt.show()
```

```python
#
# Estimación no paramétrica básica usando 
# kdeplot(). Note que aca se pasa directamente
# la columna con los datos.
#
sns.kdeplot(
    penguins.flipper_length_mm,
    color="blue",
    shade=True,
)

plt.show()
```

```python
#
# kde acumulada usando displot().
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    color='red',
    fill=True,
    cumulative=True,
)
plt.show()
```

```python
#
# kde acumulada usando kdeplot()
#
sns.kdeplot(
    penguins.flipper_length_mm,
    color="blue",
    shade=True,
    cumulative=True,
)

plt.show()
```

```python
#
# Modificación del ancho de banda para la 
# estimación de los kernels. Compare el 
# resultado con el gráfico siguiente.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=0.25,
)
plt.show()
```

```python
#
# Compare el resultado con el gráfico anterior.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="kde",
    bw_adjust=2,
)
plt.show()
```

```python
#
# kde por variable categórica.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    fill=True,
)
plt.show()
```

```python
#
# Efecto de apilamiento.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="kde",
    multiple="stack",
)
plt.show()
```

## Distribuciones empíricas acumuladas

```python
#
# Gráfico básico.
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    kind="ecdf",
)
plt.show()
```

```python
#
# Gráfico básico por variable categórica
#
sns.displot(
    penguins,
    x="flipper_length_mm",
    hue="species",
    kind="ecdf",
)
plt.show()
```
