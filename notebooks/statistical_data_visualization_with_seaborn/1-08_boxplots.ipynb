## Distribución de observarciones dentro de categorias

```python
#
# Gráfica de boxplots por categoria.
#
sns.catplot(
    x="day",
    y="total_bill",
    kind="box",
    data=tips,
    order=[
        'Sun',
        'Sat',
        'Fri',
        'Thur',
    ]
)

plt.show()
```

```python
#
# Gráfica de boxplots por categoria sin outliers.
#
sns.catplot(
    x="day",
    y="total_bill",
    kind="box",
    data=tips,
    sym="",
)

plt.show()
```

```python
#
# Tamaño de las barras. Por defecto se extienden
# a 1.5 * IQR. En el siguiente ejemplo se fijan
# en el 5% y el 95%.
#
sns.catplot(
    x="day",
    y="total_bill",
    kind="box",
    data=tips,
    whis=[5, 95],
)

plt.show()
```

```python
#
# Gráfica con separación por una segunda categoría.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="smoker",
    kind="box",
    data=tips,
)

plt.show()
```

```python
#
# Separación por categorías
#
g = sns.catplot(
    x="fare",
    y="survived",
    row="class",
    kind="box",
    orient="h",
    height=1.5,
    aspect=4,
    data=titanic.query("fare > 0"),
)
g.set(xscale="log")
plt.show()
```

```python
#
# Box plot básico
#
sns.boxplot(
    x=penguins.flipper_length_mm,
    color='red',
)

plt.show()
```

```python
#
# Modificación de las propiedades
#
p = sns.boxplot(
    x=penguins.flipper_length_mm,
    color='lightblue',
)

for e in p.artists:
    r, g, b, a = e.get_facecolor()
    p.set_facecolor((r, g, b, 0.0))

for i, artist in enumerate(p.artists):
    artist.set_edgecolor("blue")

for j in range(0, 6):
    line = p.lines[j]
    line.set_color("red")
```
