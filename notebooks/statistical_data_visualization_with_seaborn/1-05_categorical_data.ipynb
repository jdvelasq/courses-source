```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Dataset utilizado

```python
tips = sns.load_dataset("tips")

display(
    tips.head(),
    tips.tail(),
    tips.size,
)
```

## Gráficos de conteo

```python
#
# Gráfico básico de conteo por categoría
#
sns.countplot(
    x="smoker",
    data=tips,
)

plt.show()
```

```python
#
# Separación por una categoría adicional
#
palette = {
    "Female": "blue",
    "Male": "red",
}

sns.countplot(
    x="smoker",
    hue="sex",
    palette=palette,
    data=tips,
)
plt.show()
```

```python
#
# Gráfico de barras
#
sns.barplot(
    x="size",
    y="total_bill",
    data=tips,
    errwidth=0,
)

plt.show()
```

## Gráficos categóricos de puntos 

```python
#
# Gráfico básico de puntos por categoría.
#
sns.catplot(
    x="day",
    y="total_bill",
    data=tips,
)

plt.show()
```

```python
#
# Efecto de jitter
#
sns.catplot(
    x="day",
    y="total_bill",
    jitter=False,
    data=tips,
)

plt.show()
```

```python
#
# Gráfica kind="swarm" con separación por
# categorías.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# Gráfica para variables numéricas categóricas,
# modificando el orden del eje x.
#
sns.catplot(
    x="size",
    y="total_bill",
    data=tips,
    order=[2, 1, 3, 4, 5, 6],
)

plt.show()
```

```python
#
# La gráfica puede girarse con solo invertir las
# variables en los parámetros
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="time",
    kind="swarm",
    data=tips,
)

plt.show()
```

## Distribución de observarciones dentro de categorias

```python
#
# Gráfica de boxplots por categoria.
#
sns.catplot(
    x="day",
    y="total_bill",
    kind="box",
    data=tips,
    order=[
        'Sun',
        'Sat',
        'Fri',
        'Thur',
    ]
)

plt.show()
```

```python
#
# Gráfica de boxplots por categoria sin outliers.
#
sns.catplot(
    x="day",
    y="total_bill",
    kind="box",
    data=tips,
    sym="",
)

plt.show()
```

```python
#
# Tamaño de las barras. Por defecto se extienden
# a 1.5 * IQR. En el siguiente ejemplo se fijan
# en el 5% y el 95%.
#
sns.catplot(
    x="day",
    y="total_bill",
    kind="box",
    data=tips,
    whis=[5, 95],
)

plt.show()
```

```python
#
# Gráfica con separación por una segunda categoría.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="smoker",
    kind="box",
    data=tips,
)

plt.show()
```

```python
#
# Gráfica básica de violin por categoría.
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="sex",
    kind="violin",
    data=tips,
)

plt.show()
```

```python
#
# Modificación de los parámetros.
#
sns.catplot(
    x="total_bill",
    y="day",
    hue="sex",
    kind="violin",
    bw=0.15,
    cut=0,
    data=tips,
)

plt.show()
```

```python
#
# Construcción de una gráfica comparativa usando
# split.
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="violin",
    split=True,
    data=tips,
)

plt.show()
```

```python
#
# Modificación de la visualización de los datos
# al interior de las figuras
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="sex",
    kind="violin",
    inner="stick",
    split=True,
    palette="pastel",
    data=tips,
)

plt.show()
```

```python
#
# Combinación de swarm con violin
#
g = sns.catplot(
    x="day",
    y="total_bill",
    kind="violin",
    inner=None,
    data=tips,
)

sns.swarmplot(
    x="day",
    y="total_bill",
    color="k",
    size=3,
    data=tips,
    ax=g.ax,
)

plt.show()
```

## Estimación estadística dentro de categorías

```python
titanic = sns.load_dataset("titanic")

display(
    titanic.head(),
    titanic.tail(),
    titanic.size,
)
```

```python
#
# Gráfico de barras con estimados de los
# intervalos de confianza por categoría usando
# bootstrap
#
sns.catplot(
    x="sex",
    y="survived",
    hue="class",
    kind="bar",
    data=titanic,
)

plt.show()
```

```python
#
# Gráfico de barras horizontales
#
sns.catplot(
    x="survived",
    y="sex",
    hue="class",
    kind="bar",
    data=titanic,
)

plt.show()
```

```python
#
# Modificación de la paleta de colores
sns.catplot(
    x="deck",
    kind="count",
    palette="ch:.25",
    data=titanic,
)

plt.show()
```

```python
#
# Muchas de las opciones ya abordadas funcionan
# con estos gráficos
#
sns.catplot(
    y="deck",
    hue="class",
    kind="count",
    palette="pastel",
    edgecolor=".6",
    data=titanic,
)

plt.show()
```

## Gráficos de punto

```python
#
# Vista de los estimados para cada clase. Las 
# barras cubren el intervalo del 95% de confianza
# y la media
#
sns.catplot(
    x="sex",
    y="survived",
    hue="class",
    kind="point",
    data=titanic,
    capsize=0.1,
)

plt.show()
```

```python
#
# Sin invervalos de confianza
#
sns.catplot(
    x="sex",
    y="survived",
    hue="class",
    kind="point",
    data=titanic,
    ci=None,
)

plt.show()
```

```python
#
# Modificación de las propiedades del gráfico
#
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic,)

plt.show()
```

## Visualización de relaciones múltiples

```python
#
# Separación por categorías
#
sns.catplot(
    x="day",
    y="total_bill",
    hue="smoker",
    col="time",
    aspect=0.7,
    kind="swarm",
    data=tips,
)

plt.show()
```

```python
#
# Separación por categorías
#
g = sns.catplot(
    x="fare",
    y="survived",
    row="class",
    kind="box",
    orient="h",
    height=1.5,
    aspect=4,
    data=titanic.query("fare > 0"),
)
g.set(xscale="log")
plt.show()
```
