Aprendizaje Minibatch
===

## Definición del problema de estimación de parámetros

Se desean encontra los parámetros $w_0$ y $w_1$ del siguiente modelo de regresión entre las variables $x$ y $y$:

$$y_i = w_1 x_i + w_0 + e_i$$

a partir de un conjunto de observaciones $\{d_i, x_i\}$, donde $d_i$ es el valor real (observado), $y_i$ es su aproximación usando el modelo de regresión descrito, y $e_i$ es el error.

## Funciones de pérdida

* Error instantáneo:

$$
e_i = d_i - y_i
$$

* Sumatoria del error cuadrático instantáneo:

$$
\begin{split}
\text{SSE}(w_0, w_1) & = \sum_p (d_i - y_i)^2 \\
    & =  \sum_p e_i^2
\end{split}
$$


## Método del gradiente

* Aprendizaje Online:

$$
\mathbf{w}_k = \mathbf{w}_{k-1} - \mu \left( \frac{\partial}{\partial \mathbf{w}}  e_i^2 \right)
$$


* Aprendizaje Batch:

$$
\begin{split}
\mathbf{w}_k 
  & = \mathbf{w}_{k-1} - \mu \sum_i  \left( \frac{\partial}{\partial \mathbf{w}} e_i^2 \right)
\end{split}
$$

## Minibatch

Para conjuntos de datos muy grandes el aprendizaje batch es muy costoso computacionalmente ya que deben recorrerse todos los datos, mientras que el aprendizaje online es muy lento en converger . Es en este caso, donde se toma un punto intermedio: 

* El conjunto de datos se divide en varios subconjuntos.

* Se ordenan aleatoriamente los subconjuntos de datos y luego se recorren uno a uno, tal como en el caso del gradiente descendente estocástico.

* Para el subconjunto actual, se computa el gradiente neto (tal como en el caso del aprendizaje batch) y se corrigen los parámetros.

* Se repite nuevamente el proceso hasta que el algoritmo converga.

## Implementación

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
#
# Dataset
# 
x_values = [
    0.1087, 0.2698, 0.3765, 0.2146, 0.9155,
    0.0246, 0.0221, 0.8632, 0.6460, 0.2092,
    0.8567, 0.1591, 0.9647, 0.6231, 0.7460,
    0.3654, 0.3065, 0.6886, 0.4966, 0.2008,
    0.2618, 0.7607, 0.1563, 0.4424, 0.7731,
]

d_values = [
    0.9519, 1.1237, 1.2360, 1.0526, 2.0743,
    0.7906, 0.7603, 2.0533, 1.6887, 1.0563,
    2.0991, 0.8953, 2.1917, 1.6266, 1.8508,
    1.2828, 1.2283, 1.8722, 1.4657, 1.0418,
    1.1097, 1.7826, 0.9711, 1.4267, 1.8248,
]
```

```python
#
# Función auxiliar para computar el SSE
#
def compute_SSE(w0, w1):
    return sum(
        [
            (d_value - w0 - w1 * x_value) ** 2
            for x_value, d_value in zip(x_values, d_values)
        ]
    )
```

```python
#
# Función auxililar para graficar el contorno
#
def plot_contour():

    W0 = np.linspace(0.0, 2.0, 50)
    W1 = np.linspace(1.5, 3.2, 50)
    W0, W1 = np.meshgrid(W0, W1)
    F = compute_SSE(W0, W1)
    
    levels = [0, 0.2, 0.5, 0.75, 1, 2, 3, 5, 10, 20, 40, 60, 80, 100, 120]

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect("equal", "box")

    ax.contourf(
        W0,
        W1,
        F,
        cmap=cm.Greys,
        levels=levels,
        alpha=0.8,
    )

    ax.contour(
        W0,
        W1,
        F,
        colors="gray",
        levels=levels,
    )
    ax.grid()
    
    ax.set_xlabel('w0')
    ax.set_ylabel('w1')
```

```python
#
# Esta es el mismo computo que el empleado en el
# método batch.
#
def compute_batch_gradient_SSE(w0, w1, x_batch, d_batch):

    instant_errors = [
        (d_value - w0 - w1 * x_value) for x_value, d_value in zip(x_batch, d_batch)
    ]

    net_gradient_SSE_w0 = sum([-2 * error for error in instant_errors])

    net_gradient_SSE_w1 = sum(
        [-2 * error * x_value for (error, x_value) in zip(instant_errors, x_batch)]
    )

    return net_gradient_SSE_w0, net_gradient_SSE_w1
```

```python
def compute_iteration(w0, w1, learning_rate, x_batch, d_batch):

    net_gradient_SSE_w0, net_gradient_SSE_w1 = compute_batch_gradient_SSE(
        w0, w1, x_batch, d_batch
    )

    w0 = w0 - learning_rate * net_gradient_SSE_w0
    w1 = w1 - learning_rate * net_gradient_SSE_w1

    return w0, w1
```

```python
#
# Particionamiento en grupos
#
n_groups = 5
groups = np.array_split(np.array(range(len(x_values))), n_groups)
groups
```

```python
#
# Punto de inicio
#
w0 = 0.5
w1 = 3.0

history_w0 = [w0]
history_w1 = [w1]
history_f = [compute_SSE(w0, w1)]

learning_rate = 0.05

x_values = np.array(x_values)
d_values = np.array(d_values)

for epoch in range(10):
    
    
    for i_group in np.random.choice(n_groups, n_groups, replace=False):
                    
        indexes = list(groups[i_group])
        
        x_batch = x_values[indexes]
        d_batch = d_values[indexes]
        
        w0, w1 = compute_iteration(w0, w1, learning_rate, x_batch, d_batch)
        history_w0.append(w0)
        history_w1.append(w1)
        history_f.append(compute_SSE(w0, w1))

print(" w0 = {:6.4f}\n w1 = {:6.4f}\nSSE = {:6.4f}\n".format(w0, w1, compute_SSE(w0, w1)))

plot_contour()
plt.plot(history_w0, history_w1, marker='o', color="black");
```

```python
plt.figure(figsize=(10,6))
plt.plot(history_f, ".-k")
plt.xlabel('Iteraciones')
plt.ylabel('SSE(w0, w1)')
plt.show()
```
