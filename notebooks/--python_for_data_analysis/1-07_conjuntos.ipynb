Un conjunto es una estructura de datos cuyos elementos no se repiten.

```python
#
# Un conjunto es una estructura de datos 
# cuyos elementos no se repiten
#
set_a = { 'a', 'a', 'b', 'c', 'd', 'd', 'd'}
set_a
```

## Métodos

```python
#
# add()
# ===================================
#
set_a = {'a', 'b', 'c', 'd'}
set_a.add(1)
set_a
```

```python
#
# discard()
# ===================================
#
set_a = {'a', 'b', 'c', 'd'}
set_a.discard('a')
set_a
```

```python
#
# pop()
# ===================================
#
set_a = {'a', 'b', 'c', 'd'}
set_a.pop()
set_a
```

```python
#
# update()
# ===================================
#
set_a = {'a', 'b', 'c', 'd'}
set_b = {1, 2, 3, 4}
set_a.update(set_b)
set_a
```

```python
#
# difference
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a.difference(set_b)
```

```python
#
# union
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a.union(set_b)
```

```python
#
# intersection
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a.intersection(set_b)
```

## Operadores

```python
#
# Diferencia de conjuntos
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a - set_b                              
```

```python
#
# Unión de conjuntos
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a | set_b 
```

```python
#
# Intersección de conjuntos
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a & set_b
```

```python
#
# Unión menos intersección
# ===================================
#
set_a = {0, 1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7, 8}
set_a ^ set_b
```

```python
#
# Pertenencia
# ===================================
#
'a' in set_a
```

## Ejemplo

A partir del archivo babe_names.csv, encuentre los nombres que aparecen en 2014 y no en 2011.

```python
babynames_url = "https://raw.githubusercontent.com/jdvelasq/datalabs/master/datasets/baby_names.csv" 
!wget --quiet {babynames_url} -P /tmp/ 
```

```python
#
# El detalle de este codigo es explicado más 
# adelante
#
import pandas as pd

babynames_df = pd.read_csv("/tmp/baby_names.csv")
babynames_df.head()
```

```python
#
# Listas con los nombres para los años 2011 y 2014
#
babynames_2011_list = babynames_df[babynames_df.BRITH_YEAR == 2011].NAME.tolist()
babynames_2014_list = babynames_df[babynames_df.BRITH_YEAR == 2014].NAME.tolist()
```

```python
babynames_2011_set = set(babynames_2011_list)
babynames_2014_set = set(babynames_2014_list)
babynames_2014_set - babynames_2011_set
```
