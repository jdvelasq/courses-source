```python
#
# Carga de datos del iris flowers
#
from sklearn import datasets

data = datasets.load_iris()
X, y = data.data, data.target
X.shape, y.shape
```

```python
#
# Distribución de los datos por atributo
#
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(10, 6))
sns.boxplot(data=pd.DataFrame(X, columns=data.feature_names))
plt.xticks(rotation=90);
```

```python
#
# Escalamiento
#
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_norm = scaler.fit_transform(X)

plt.figure(figsize=(10, 6))
sns.boxplot(data=pd.DataFrame(X_norm, columns=data.feature_names))
plt.xticks(rotation=90)
plt.show()
```

```python
#
# Dimensiones del SOM
#
from som import SOM

NROWS = 30
NCOLS = 30

som = SOM(nrows=NROWS, ncols=NCOLS, ndim=4, max_iter=100)
som.fit(X_norm)

#
# U matrix: 
#
som.plot_Umatrix()
```

```python
#
# Reducción de la dimensionalidad. Cada color representa
# una clase
#
som.dimensionality_reduction(X_norm, y)
```
