```python
df_1 = pd.read_csv("/tmp/data_1.csv")
df_2 = pd.read_csv("/tmp/data_2.csv")
display(df_1, df_2)
```

```python
#
# Sufijos generados para desambiguar los nombres
# de las columnas
#
pd.merge(
    df_1,
    df_2,
    on="clientId",
)
```

```python
#
# Especificación del sufijo
#
pd.merge(
    df_1,
    df_2,
    on="clientId",
    suffixes=["_LD", "_RD"],
)
```

```python
#
# Renombramiento de las columnas
#
pd.merge(
    df_1,
    df_2,
    on="clientId",
).rename(
    columns={
        "info_x": "name",
        "info_y": "location",
    }
)
```
