Escala los valores en el rango [-1, +1], dividiendo por el valor máximo de cada característica.

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")

data = penguins[["flipper_length_mm"]]
data = data.rename(columns={"flipper_length_mm": "original"})
```

```python
from sklearn.preprocessing import MaxAbsScaler

maxAbsScaler = MaxAbsScaler()
maxAbsScaler.fit(data)

data["transformed"] = maxAbsScaler.transform(data)
```

```python
import matplotlib.pyplot as plt

sns.jointplot(x="original", y="transformed", data=data, kind="scatter")
plt.show()
```
