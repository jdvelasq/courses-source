Gráfico de líneas en Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")

groupby_timepoint = fmri.groupby(["event", "timepoint"]).describe()


display(
    groupby_timepoint.head(),
    groupby_timepoint.tail(),
    groupby_timepoint.size
)
```

```python
#
# Grafica la serie para 'stim'
#
plt.plot(
    groupby_timepoint.loc['stim']['signal']['mean'],
    color='tab:blue',
    linewidth=2,
    label='stim',
)

plt.fill_between(
    x = groupby_timepoint.loc['stim'].index,
    y1=groupby_timepoint.loc['stim']['signal']['25%'],
    y2=groupby_timepoint.loc['stim']['signal']['75%'],
    color='tab:blue',
    alpha=0.15,
)

#
# Grafica la serie para 'cue'
#
plt.plot(
    groupby_timepoint.loc['cue']['signal']['mean'],
    color='tab:orange',
    linewidth=2,
    label='cue',
)

plt.fill_between(
    x = groupby_timepoint.loc['cue'].index,
    y1=groupby_timepoint.loc['cue']['signal']['25%'],
    y2=groupby_timepoint.loc['cue']['signal']['75%'],
    color='tab:orange',
    alpha=0.15,
)


#
# Estética
#
plt.legend()
plt.gca().set_aspect("auto")

plt.gca().spines["left"].set_color("gray")
plt.gca().spines["bottom"].set_color("gray")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```
