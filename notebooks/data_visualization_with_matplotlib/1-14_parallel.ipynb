```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Coordinadas paralelas en Matplotlib

```python
from pandas.plotting import parallel_coordinates

iris = sns.load_dataset('iris')

parallel_coordinates(
    iris, 
    'species', 
)

plt.show()
```

## Gráficos de punto en Seaborn

```python
titanic = sns.load_dataset('titanic')

#
# Vista de los estimados para cada clase. Las 
# barras cubren el intervalo del 95% de confianza
# y la media
#
sns.catplot(
    x="sex",
    y="survived",
    hue="class",
    kind="point",
    data=titanic,
    capsize=0.1,
)

plt.show()
```

```python
#
# Sin invervalos de confianza
#
sns.catplot(
    x="sex",
    y="survived",
    hue="class",
    kind="point",
    data=titanic,
    ci=None,
)

plt.show()
```

```python
#
# Modificación de las propiedades del gráfico
#
sns.catplot(
    x="class",
    y="survived",
    hue="sex",
    palette={"male": "g", "female": "m"},
    markers=["^", "o"],
    linestyles=["-", "--"],
    kind="point",
    data=titanic,
)

plt.show()
```
