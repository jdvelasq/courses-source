```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes

db = load_diabetes()
df = pd.DataFrame(db.data)
df.columns = db.feature_names
df['target'] = db.target
df.head()
```

```python
df.isnull().sum()
```

```python
df['age'][3] = np.nan
df.head()
```

```python
df.isnull().mean()
```

## Imputación

```python
df.columns[df.isnull().mean() > 0]
```

```python
threshold = 0.001

# Borrado de columnas con una tasa de valores faltantes mayor 
# que el threshold
data = df[df.columns[df.isnull().mean() < threshold]]
data.head()
```

```python
#
# Llenado de faltantes con cero
#
data = df.fillna(0)
data.head()
```

```python
data = df.fillna(df.median())
data.head()
```

```python
from sklearn.impute import SimpleImputer

i_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
i_mean.fit(df)

display(f'Mean {df["age"].mean()}')  

imputed = pd.DataFrame(i_mean.transform(df))
imputed.head()
```

## Outliers

```python
#
# Criterio de la desviación estándar
#
upper = df['age'].mean() + 3 * df['age'].std()
lower = df['age'].mean() - 3 * df['age'].std()
data = df[ (df['age'] < upper) & (df['age'] > lower)]
data.head()
```

```python
#
# Criterio de los percentiles
#
iq_range = upper - lower

data = df[
    (df['age'] < upper + 1.5 * iq_range) & (df['age'] > lower - 1.5 * iq_range)
]

data.head()
```

## Binning

```python
df['binned_age'] = pd.cut(
    df['age'],
    bins=[0, 0.01, 0.02, 0.05, 0.09],
    labels=['low', 'mid-low', 'mid-high', 'high'],
)

df.head()
```
