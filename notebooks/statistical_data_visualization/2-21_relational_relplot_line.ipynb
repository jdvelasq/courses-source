Gráficos básicos de línea con relplot() ----

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#
# Nuevo dataset de ejemplo
#
dots = sns.load_dataset("dots").query("align == 'dots'")

display(
    dots.head(),
    dots.tail(),
    dots.size
)
```

```python
#
# Color basado en una columna numérica.
#
sns.relplot(
    x="time",
    y="firing_rate",
    hue="coherence",
    kind="line",
    data=dots,
    ci=None,
)

plt.show()
```

```python
#
# Color basado en una columna numérica y separación por clases. Note que el
# esquema de colores es lineal y no permite visualizar bien las diferencias.
#
sns.relplot(
    x="time",
    y="firing_rate",
    hue="coherence",
    style="choice",
    kind="line",
    data=dots,
    ci=None,
)
plt.show()
```

```python
#
# Definición de una escala logaritmica para los colores
#
palette = sns.cubehelix_palette(
    light=0.8,
    n_colors=6,
)

sns.relplot(
    x="time",
    y="firing_rate",
    hue="coherence",
    style="choice",
    palette=palette,
    kind="line",
    data=dots,
)

plt.show()
```

```python
#
# Alternativa: modificación de la forma de normalizar la paleta de colores
#
from matplotlib.colors import LogNorm

palette = sns.cubehelix_palette(
    light=0.7,
    n_colors=6,
)

sns.relplot(
    x="time",
    y="firing_rate",
    hue="coherence",
    style="choice",
    hue_norm=LogNorm(),
    kind="line",
    data=dots.query("coherence > 0"),
)

plt.show()
```

```python
#
# Cambio del grosor de las líneas usando una columna numérica
#
sns.relplot(
    x="time",
    y="firing_rate",
    size="coherence",
    style="choice",
    kind="line",
    data=dots,
)

plt.show()
```

```python
#
# Cambio del grosor de las líneas usando una columna categórica
#
sns.relplot(
    x="time",
    y="firing_rate",
    hue="coherence",
    size="choice",
    palette=palette,
    kind="line",
    data=dots,
)
plt.show()
```
