Gráficos de scatter con relplot() ----
===

* 0:00 min | Última modificación: Octubre 13, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#
# Gráfico básico
#
sns.color_palette("pastel")

# (Figure-level function)
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    color='tab:green',
    kind='scatter',
)

# (Axes-level function)
sns.rugplot(
    data=tips,
    x="total_bill",
    y="tip",
    color='tab:green',
)

plt.show()
```

```python
#
# Separación de categoría por color
# =============================================================================
#
# (Figure-level function)
sns.relplot(
    x="total_bill",
    y="tip",
    hue="smoker",
    data=tips,
    kind='scatter',
)

# (Axes-level function)
g = sns.rugplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
)

g.legend_.remove()

plt.show()
```

```python
#
# Separación de categoría por forma
#
sns.relplot(
    x="total_bill",
    y="tip",
    style="smoker",
    data=tips,
    kind='scatter',
)
plt.show()
```

```python
#
# Separación de categoría por size
#
sns.relplot(
    x="total_bill",
    y="tip",
    size="smoker",
    sizes=(50, 200),
    data=tips,
    alpha=0.4,
    kind='scatter',
)
plt.show()
```

```python
#
# Separación de categoría por gráfica
#
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    col='day',
    kind='scatter',
)

plt.show()
```

```python
#
# Separación de categoría por gráfica.
# Modificación de la cantidad de columnas por fila y cambio en el orden.
#
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    kind='scatter',
    col='day',
    col_wrap=2,
    col_order = [
        'Thur',
        'Fri',
        'Sat',
        'Sun',
    ],
)

plt.show()
```

```python
#
# Separación de variables categoría por fila y columna
#
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    col='day',
    row='time',
    kind='scatter',
)

plt.show()
```
