Fundamentos de pronóstico de series de tiempo 
===

**Serie de tiempo.** Se define como la secuencia de observaciones:

$$y_0, y_1, y_2,...,y_T$$

**Media.** 

$$\mu_t = E(y_t)$$

**Covarianza.**

$$\gamma_{t, s} = Cov(y_t, y_s)$$

donde: $Cov(y_t, y_s) = E(y_t - \mu_t)(y_s - \mu_s) = E(y_t * y_s) - \mu_t \mu_s$ 

**Función de autocorrelación.**

$$\rho_{t, s} = \frac{Cov(y_t, y_s)}{\sqrt{Var(y_t) Var(y_s)}}$$
