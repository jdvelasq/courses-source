```python
dataset_1 = pd.read_csv("/tmp/dataset_1.csv")
dataset_2 = pd.read_csv("/tmp/dataset_2.csv")
datasets_12 = pd.concat(
    [
        dataset_1,
        dataset_2,
    ]
)
datasets_12.head()
```
