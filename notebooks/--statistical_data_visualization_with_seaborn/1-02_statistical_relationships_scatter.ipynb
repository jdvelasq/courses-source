Visualización de relaciones estadísticas con Scatter Plots
===

* 60:00 min | Última modificación: Abril 5, 2021 | [YouTube]

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Carga del dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Construcción de los gráficos

```python
#
# Gráfico básico
# ===============================================
#
sns.color_palette("pastel")

sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
)
plt.show()
```

```python
#
# Separación de categoría por color
# ===============================================
#
sns.relplot(
    x="total_bill",
    y="tip",
    hue="smoker",
    data=tips,
)
plt.show()
```

```python
#
# Separación de categoría por forma
# ===============================================
#
sns.relplot(
    x="total_bill",
    y="tip",
    style="smoker",
    data=tips,
)
plt.show()
```

```python
#
# Separación de categoría por size
# ===============================================
#
sns.relplot(
    x="total_bill",
    y="tip",
    size="smoker",
    sizes=(50, 200),
    data=tips,
    alpha=0.4,
)
plt.show()
```

```python
#
# Separación de categoría por gráfica
# ===============================================
#
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    col='day',
)

plt.show()
```

```python
#
# Separación de categoría por gráfica
# ===============================================
# Modificación de la cantidad de columnas por
# fila y cambio en el orden
#
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    kind='scatter',
    col='day',
    col_wrap=2,
    col_order = [
        'Thur',
        'Fri',
        'Sat',
        'Sun',
    ]
)

plt.show()
```

```python
#
# Separación de variables categoría por fila y
# columna
# ===============================================
#
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    col='day',
    row='time',
)

plt.show()
```
